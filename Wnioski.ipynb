# Optymalizacja Hiperparametrów

Ogólnie rzec ujmując nikt w tym zadaniu nie optymalizował hiperparametrów (przeszukałem +-5 prac).

Dlaczego? Prace na kagglu z tego co zawuażyłem często charakteryzują się mocnym skupieniem na wyniku, a nie procesie, który często jest kastrowany w wersji wrzucanej do konkursu.

Optymalnie jednak warto było zrobić przeszukiwanie idealnych hiperparametrów GridSearchem

Dowiedziałem się natomiast, żę najlepszych pracach na Kaggle natomiast stosuje się optymalizację Bayesowską. (Biblioteka Optuna w pythonie). Algorytm uczy się w niej, które parametry dają lepsze wyniki i w kolejnych krokach sprawdza tylko te obiecujące obszary.
